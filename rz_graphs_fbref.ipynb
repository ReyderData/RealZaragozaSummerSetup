```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_html('https://fbref.com/en/comps/17/misc/Segunda-Division-Stats', 
                  attrs={'id': 'stats_squads_misc_for'})[0]
df
```

```python
df = df[['Unnamed: 0_level_0', 'Performance']].copy()
df.columns = df.columns.droplevel()  # drop the top-level of the multi-index
df
```

```python
df = df.drop(["CrdR", "2CrdY", "PKwon", "PKcon"], axis = 1) 
df
```

```python
df['FlsPerYc'] = (df['Fls']/df['CrdY']).round(2)
df['DefAct'] = (df['Int']+df['TklW'])
df
```

```python
#Filtering rows
label = 'Zaragoza'
row_label = df[df['Squad'] == label]
other_rows = df[df['Squad'] != label]

fig, ax = plt.subplots(figsize=(10, 6))

# Scatter other teams
ax.scatter(x=other_rows['Fls'], y=other_rows['FlsPerYc'],
           color='lightgray', edgecolor='black', alpha=0.6, s=60)

# Scatter Zaragoza
ax.scatter(x=row_label['Fls'], y=row_label['FlsPerYc'],
           color='dodgerblue', edgecolor='black', s=150, label=label, zorder=3)

# Zaragoza label
for i, row in row_label.iterrows():
    ax.annotate(f"{row['Squad']}\n({row['Fls']}, {row['FlsPerYc']:.2f})",
                (row['Fls'], row['FlsPerYc']),
                textcoords='offset points', xytext=(10, -10),
                ha='left', fontsize=10, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', fc='white', ec='dodgerblue', lw=1))

# Axes Customization
ax.set_xlabel("Faltas", fontsize=12)
ax.set_ylabel("Faltas por T. Amarilla", fontsize=12)
ax.invert_yaxis() 
ax.set_title("Relación de faltas y nº de faltas por tarjeta amarilla", fontsize=12, fontweight='bold')


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
```

```python
#Sort the df by selected stat
df_sorted = df.sort_values(by='DefAct', ascending=False).reset_index(drop=True)

# Select color blue just for labelled team
colors = ['lightblue' if team == label else 'gray' for team in df_sorted['Squad']]

# Plot base
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Lineplot
plt.plot(df_sorted['Squad'], df_sorted['DefAct'], color='lightgray', linewidth=2, zorder=1)

# Scatter points for each team
plt.scatter(df_sorted['Squad'], df_sorted['DefAct'], color=colors, s=80, edgecolor='black', zorder=2)

# Label for desired team
highlight = df_sorted[df_sorted['Squad'] == label]
for i, row in highlight.iterrows():
    plt.annotate(f"{row['Squad']}: {row['DefAct']:.1f}",
                 (row['Squad'], row['DefAct']),
                 textcoords='offset points', xytext=(0, 10), ha='center',
                 fontsize=10, fontweight='bold',
                 bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="lightblue", lw=1.5))

# plt customization
plt.title("Acciones Defensivas (Tckls + Ints)", fontsize=16, fontweight='bold')
plt.ylabel("DefAct", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.3)
sns.despine()
plt.tight_layout()
```

```python
teams = pd.read_csv('laliga2_att_stats.csv')
teams.head(2)
```

```python
# Filtering all numeric cols
numeric_cols = teams.select_dtypes(include='number').columns

# Get desired team
zaragoza_row = teams[teams['Squad'] == 'Zaragoza']

# Config desired cols & rows structure
n_cols = 5
n_rows = -(-len(numeric_cols) // n_cols) 
plt.figure(figsize=(25, 10))

# Boxplot for each stat
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(data=teams[col], orient='h', color='lightblue')
    
    zaragoza_value = zaragoza_row[col].values[0]
    plt.axvline(zaragoza_value, color='red', linestyle='--', linewidth=2)
    
    plt.title(f"{col}\nZaragoza: {zaragoza_value:.2f}", fontsize=12, fontweight='bold')
    plt.xlabel("")

plt.tight_layout()
```
